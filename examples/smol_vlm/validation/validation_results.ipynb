```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib.patches import Rectangle

# Set style for better-looking plots
plt.style.use('default')
sns.set_palette("husl")
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD']

# Create figure with subplots
fig = plt.figure(figsize=(20, 16))
gs = fig.add_gridspec(4, 3, height_ratios=[1, 1.2, 1, 1], width_ratios=[1, 1, 1])

# 1. Overall Benchmark Comparison
ax1 = fig.add_subplot(gs[0, :])
benchmarks = ['MMStar', 'CCBench', 'AI2D_TEST', 'SEEDBench2_Plus']
kornia_scores = [0.398, 0.188, 0.641, 0.542]
transformers_scores = [0.417, 0.206, 0.645, 0.573]

x = np.arange(len(benchmarks))
width = 0.35

bars1 = ax1.bar(x - width/2, kornia_scores, width, label='Kornia (Rust)', color=colors[0], alpha=0.8)
bars2 = ax1.bar(x + width/2, transformers_scores, width, label='Transformers (Python)', color=colors[1], alpha=0.8)

ax1.set_xlabel('Benchmarks', fontsize=12, fontweight='bold')
ax1.set_ylabel('Score', fontsize=12, fontweight='bold')
ax1.set_title('VLM Performance Comparison: Kornia vs Transformers', fontsize=16, fontweight='bold', pad=20)
ax1.set_xticks(x)
ax1.set_xticklabels(benchmarks)
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)

# Add value labels on bars
for bar in bars1:
    height = bar.get_height()
    ax1.annotate(f'{height:.3f}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=10)
for bar in bars2:
    height = bar.get_height()
    ax1.annotate(f'{height:.3f}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=10)

# 2. MME Detailed Breakdown
ax2 = fig.add_subplot(gs[1, :])
mme_categories = ['Perception', 'Reasoning', 'OCR', 'Artwork', 'Celebrity', 'Code_reasoning', 
                  'Color', 'Commonsense_reasoning', 'Count', 'Existence', 'Landmark', 
                  'Numerical_calculation', 'Position', 'Posters', 'Scene', 'Text_translation']
kornia_mme = [1347.6, 240.714, 132.5, 132, 87.9412, 50, 168.333, 95.7143, 143.333, 
              190, 138, 47.5, 80, 112.245, 163.25, 47.5]
transformers_mme = [1364.56, 284.286, 132.5, 126.25, 88.8235, 62.5, 180, 99.2857, 140, 
                    195, 134, 50, 90, 115.986, 162, 72.5]

x_mme = np.arange(len(mme_categories))
bars1_mme = ax2.bar(x_mme - width/2, kornia_mme, width, label='Kornia (Rust)', color=colors[0], alpha=0.8)
bars2_mme = ax2.bar(x_mme + width/2, transformers_mme, width, label='Transformers (Python)', color=colors[1], alpha=0.8)

ax2.set_xlabel('MME Categories', fontsize=12, fontweight='bold')
ax2.set_ylabel('Score', fontsize=12, fontweight='bold')
ax2.set_title('MME Benchmark Detailed Breakdown', fontsize=14, fontweight='bold', pad=15)
ax2.set_xticks(x_mme)
ax2.set_xticklabels(mme_categories, rotation=45, ha='right')
ax2.legend(fontsize=11)
ax2.grid(True, alpha=0.3)

# Add value labels on MME bars
for bar in bars1_mme:
    height = bar.get_height()
    ax2.annotate(f'{height:.1f}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=8)
for bar in bars2_mme:
    height = bar.get_height()
    ax2.annotate(f'{height:.1f}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=8)

# 3. COCO_VAL Metrics Comparison
ax3 = fig.add_subplot(gs[2, 0])
coco_metrics = ['BLEU-1', 'BLEU-2', 'BLEU-3', 'BLEU-4', 'ROUGE-L', 'CIDEr']
kornia_coco = [4.977, 3.192, 1.763, 0.962, 8.224, 0.131]
transformers_coco = [4.259, 2.756, 1.513, 0.820, 6.300, 0.012]

x_coco = np.arange(len(coco_metrics))
bars1_coco = ax3.bar(x_coco - width/2, kornia_coco, width, label='Kornia', color=colors[0], alpha=0.8)
bars2_coco = ax3.bar(x_coco + width/2, transformers_coco, width, label='Transformers', color=colors[1], alpha=0.8)

ax3.set_xlabel('COCO Metrics', fontsize=11, fontweight='bold')
ax3.set_ylabel('Score', fontsize=11, fontweight='bold')
ax3.set_title('COCO_VAL Caption Metrics', fontsize=12, fontweight='bold')
ax3.set_xticks(x_coco)
ax3.set_xticklabels(coco_metrics, rotation=45, ha='right')
ax3.legend(fontsize=10)
ax3.grid(True, alpha=0.3)

# Add value labels on COCO bars
for bar in bars1_coco:
    height = bar.get_height()
    ax3.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=9)
for bar in bars2_coco:
    height = bar.get_height()
    ax3.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=9)

# 4. SEEDBench2_Plus Breakdown
ax4 = fig.add_subplot(gs[2, 1])
seed_categories = ['Overall', 'Chart', 'Map', 'Web']
kornia_seed = [0.542, 0.499, 0.462, 0.692]
transformers_seed = [0.573, 0.544, 0.466, 0.739]

x_seed = np.arange(len(seed_categories))
bars1_seed = ax4.bar(x_seed - width/2, kornia_seed, width, label='Kornia', color=colors[0], alpha=0.8)
bars2_seed = ax4.bar(x_seed + width/2, transformers_seed, width, label='Transformers', color=colors[1], alpha=0.8)

ax4.set_xlabel('SEEDBench Categories', fontsize=11, fontweight='bold')
ax4.set_ylabel('Score', fontsize=11, fontweight='bold')
ax4.set_title('SEEDBench2_Plus Breakdown', fontsize=12, fontweight='bold')
ax4.set_xticks(x_seed)
ax4.set_xticklabels(seed_categories)
ax4.legend(fontsize=10)
ax4.grid(True, alpha=0.3)

# Add value labels on SEEDBench bars
for bar in bars1_seed:
    height = bar.get_height()
    ax4.annotate(f'{height:.3f}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=9)
for bar in bars2_seed:
    height = bar.get_height()
    ax4.annotate(f'{height:.3f}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=9)

# 5. POPE Metrics Comparison
ax5 = fig.add_subplot(gs[2, 2])
pope_metrics = ['Overall', 'Accuracy', 'Precision', 'Recall']
kornia_pope = [84.69, 85.96, 93.10, 77.67]
transformers_pope = [85.43, 86.48, 92.63, 79.27]

x_pope = np.arange(len(pope_metrics))
bars1_pope = ax5.bar(x_pope - width/2, kornia_pope, width, label='Kornia', color=colors[0], alpha=0.8)
bars2_pope = ax5.bar(x_pope + width/2, transformers_pope, width, label='Transformers', color=colors[1], alpha=0.8)

ax5.set_xlabel('POPE Metrics', fontsize=11, fontweight='bold')
ax5.set_ylabel('Score (%)', fontsize=11, fontweight='bold')
ax5.set_title('POPE Evaluation Metrics', fontsize=12, fontweight='bold')
ax5.set_xticks(x_pope)
ax5.set_xticklabels(pope_metrics)
ax5.legend(fontsize=10)
ax5.grid(True, alpha=0.3)

# Add value labels on POPE bars
for bar in bars1_pope:
    height = bar.get_height()
    ax5.annotate(f'{height:.1f}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=9)
for bar in bars2_pope:
    height = bar.get_height()
    ax5.annotate(f'{height:.1f}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=9)

# 6. Performance Gap Analysis
ax6 = fig.add_subplot(gs[3, :])
benchmark_names = ['MMStar', 'CCBench', 'AI2D_TEST', 'SEEDBench2_Plus', 'POPE']
performance_gaps = [
    (0.417 - 0.398) / 0.417 * 100,  # MMStar
    (0.206 - 0.188) / 0.206 * 100,  # CCBench
    (0.645 - 0.641) / 0.645 * 100,  # AI2D_TEST
    (0.573 - 0.542) / 0.573 * 100,  # SEEDBench2_Plus
    (85.43 - 84.69) / 85.43 * 100   # POPE
]

colors_gap = ['red' if gap > 0 else 'green' for gap in performance_gaps]
bars_gap = ax6.barh(benchmark_names, performance_gaps, color=colors_gap, alpha=0.7)

ax6.set_xlabel('Performance Gap (%)', fontsize=12, fontweight='bold')
ax6.set_title('Performance Gap: (Transformers - Kornia) / Transformers × 100%', fontsize=14, fontweight='bold')
ax6.axvline(x=0, color='black', linestyle='-', alpha=0.5)
ax6.grid(True, alpha=0.3)

# Add gap values as text
for i, (bar, gap) in enumerate(zip(bars_gap, performance_gaps)):
    ax6.text(gap + (0.1 if gap > 0 else -0.1), i, f'{gap:.1f}%', 
             va='center', ha='left' if gap > 0 else 'right', fontweight='bold')

plt.tight_layout()
plt.show()

# Summary Statistics
print("=== VLM Evaluation Summary ===")
print(f"Average Performance Gap: {np.mean(performance_gaps):.2f}%")
print(f"Kornia wins in: {sum(1 for gap in performance_gaps if gap < 0)} out of {len(performance_gaps)} benchmarks")
print(f"Largest gap (Transformers favor): {max(performance_gaps):.2f}% on {benchmark_names[np.argmax(performance_gaps)]}")
if min(performance_gaps) < 0:
    print(f"Largest gap (Kornia favor): {min(performance_gaps):.2f}% on {benchmark_names[np.argmin(performance_gaps)]}")
else:
    print("Kornia doesn't outperform on any benchmark")
```
